```python
from pathlib import Path
import os
```

```python
dataset = Path('Datasets/Images-20220614T161502Z-001/Images/')
output = 'gen_pairs/pairs.txt'
```

```python
os.makedirs(os.path.dirname(output), exist_ok=True)
```

```python
filenames = os.listdir(dataset)
filenames
```

```python
pairs = []
```

```python
for i in range(len(filenames)):
    for j in range(i + 1, len(filenames)):
        pair_name = filenames[i] + ' ' + filenames[j]
        pairs.append(pair_name)
```

```python
pairs
```

```python
with open(output, 'w') as f:
    for pair in pairs:
        f.write(pair+'\n')
```
